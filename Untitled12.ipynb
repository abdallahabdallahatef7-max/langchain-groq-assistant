```python
import getpass
import os
from langchain_groq import ChatGroq
```

```python
os.environ["GROQ_API_KEY"] = getpass.getpass("Groq API Key:")
```

```python
lang_model = ChatGroq(model="moonshotai/kimi-k2-instruct")
```

## Prompt Templating

```python
from langchain_core.prompts import ChatPromptTemplate ,SystemMessagePromptTemplate ,HumanMessagePromptTemplate
```

```python
prompt=ChatPromptTemplate.from_messages([
    SystemMessagePromptTemplate.from_template('You are an expert AI assistant with extensive knowledge about Albert Einstein. You know detailed information about his biography, historical context, scientific achievements, and theories such as relativity. Answer questions accurately and clearly.')
    ,
    HumanMessagePromptTemplate.from_template("user question: {question}")
])
```

## Output Parsing

#### Data Extraction: By default, LLMs return a complex object (like AIMessage). The StrOutputParser extracts only the content string from that object, making the output ready for printing or web display.

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
parser=StrOutputParser()
```

```python
chain=prompt | lang_model |parser
```

```python
result=chain.invoke("Who is Albert Einstein and why is he famous?")
```

```python
print(result)
```
